### operatior overloading 

```python
class A:
    def __init__(self,a):
        self.a=a
    def __add__(self,other):
        return self.a+other.a
obj1=A(5)
obj2=A(3)
obj3=A("hello")
obj4=A("Hi")

print(obj1+obj2)
print(obj3+obj4)

print(A.__add__(obj1,obj2))
print(A.__add__(obj3,obj4))

print(obj1.__add__(obj2))
print(obj3.__add__(obj4))
```

```python
class A:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __add__(self,other):
        return self.a+other.a,self.b+other.b
obj1=A(5,4)
obj2=A(3,7)
print(obj1+obj2)
```

```python
class A:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __sub__(self,other):
        return self.a-other.a,self.b-other.b
obj1=A(5,4)
obj2=A(3,7)
print(obj1-obj2)
```

```python
class A:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __truediv__(self,other):
        return self.a/other.a,self.b/other.b
obj1=A(5,4)
obj2=A(3,7)
print(obj1/obj2)
```

(+)--> __add__(self,other)
(-)--> __sub__(self,other)
(*)--> __mul__ (self,other)
(/)--> __truediv__(self,other)
(//)--> __floordiv__(self,other)\
(%)--> __mod__(self,other)
(**)--> __pow__(self,other)

## Method Resolution Order(M20) 

```python
class A:
    def rk(self):
        print("In class A")
class B(A):
    def rk(self):
        print("In class B")
class C(A):
    def rk(self):
        print("In class C")
class D(B,C):
    pass
r=D()
r.rk()

print(D.__mro__)
```

```python
class A:
    def rk(self):
        print("In class A")
class B:
    def rk(self):
        print("In class B")
    def rk1(self):
        print("In class B2")
class C(A):
    pass
class D(B):
    def rk1(self):
        print("In class D")
class GC(C,D):
    pass

obj=GC()
obj.rk1()
obj.rk()
```

```python
from  abc import ABC,abstractmethod
class shape(ABC):
    def __init__(self,shape_name):
        self.shape_name=shape_name
    @abstractmethod
    def draw(self):
        pass
    
class Circle(shape):
    def __init__(self):
        super().__init__("Circle")
        
    def draw(self):
        pass
    
    
    def draw_circle(self):
        print("Drawing",self.shape_name)
obj=Circle()
obj.draw_circle()
```

```python
class Staff:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
    def display(self):
        print("Name",self.name)
        print("Salary",self.salary)
        
        
class Teaching(Staff):
    def __init__(self,name,salary,subject):
        super().__init__(name,salary)
        self.subject=subject
    def display(self):
        super().display()
        print("subject",self.subject)
        
        
class NonTeaching(Staff):
    def __init__(self,name,salary,department):
        super().__init__(name,salary)
        self.department=department
    def display(self):
        super().display()
        print("department",self.department)
```

```python
class Student(self):
    def __init__(self,rollno,name,age,totalmarks):
        self.rollno=rollno
        self.name=name
        self.age=age
        self.totalmarks=totalmarks
        
    def display(self):
        print("rollno:",self.rollno)
        print("name:",self.name)
        print("age:",self.age)
        print("totalmarks:",self.totalmarks)
    
    def __eq__(self,other):
        if self.totalmarks==other.totalmarks:
            return True
        else:
            return False
```

```python
import turtle
import math

def draw_heart():
    screen = turtle.Screen()
    screen.setup(width=600, height=600)
    screen.bgcolor("white")
    
    t = turtle.Turtle()
    t.speed(10)
    t.color("red")
    t.fillcolor("red")
    t.begin_fill()
    
    t.left(140)
    t.forward(113)
    
    for _ in range(200):
        t.right(1)
        t.forward(1)
        
    t.left(120)
    
    for _ in range(200):
        t.right(1)
        t.forward(1)
        
    t.forward(113)
    t.end_fill()
    t.hideturtle()
    turtle.done()

draw_heart()
```
