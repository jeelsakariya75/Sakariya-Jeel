# Slidebar,Columns and Expander

```python
%%writefile 2_hello.py
import streamlit as st
st.set_page_config(page_title="hello Streamlit",
             page_icon="üåà",
             layout="centered")
st.title("Faculty profile Damo")
st.markdown("This example shows how to use ""sidebar,columns,expander""")
st.sidebar.header("profile satting")
faculty_name = st.sidebar.text_input("Faculty Name",'Tejas Thakkar')
departmant=st.sidebar.selectbox("departmant",['CE','CSE','IT','AIML'])
experience=st.sidebar.slider("years of Experience",0,40,10)
st.sidebar.markdown("___")
st.sidebar.write("you can put filghter,toggles etc in sidebar")
col1,col2=st.columns([1,2])
with col1:
    st.subheader("Basic info")
    st.write(f"**Name:**,{faculty_name}")
    st.write(f"**departmant:**,{departmant}")
    st.write(f"**experience**,{experience}years")
    
with col2:
    st.subheader("About")
    st.markdown("Use this area to show detailed information")
    
with st.expander("Show courses handled"):
    st.write("Python-1")
    st.write("DE")
with st.expander("Show Publication"):
    st.write('1.Research Paper A(2024)')
    st.write('1.Research Paper B(2025)')
    
```
